## verificar e contar pessoas de 20 a 30 , de 30 a 40, de 40 a 50 , de 50 a 60  

```python

pessoas_ate_30 = dados[dados['idade_pessoa'] <=30 ]

x = pessoas_ate_30.shape[0]

print(f'existem : {x} pessoas ate 30 anos ')
```

```python


pessoas_ate_40 = dados[(dados['idade_pessoa'] > 30) & (dados['idade_pessoa'] <= 40)]


y = pessoas_ate_40.shape[0]

print(f'Existem {y} pessoas com mais de 31 e atÃ© 40 anos.')
```

```python

pessoas_ate_50 = dados[(dados['idade_pessoa'] > 40) & (dados['idade_pessoa'] <= 50)]


a = pessoas_ate_50.shape[0]

print(f'Existem {a} pessoas de 41 e atÃ© 50 anos.')
```

```python
pessoas_ate_60 = dados[(dados['idade_pessoa'] > 50) & (dados['idade_pessoa'] <= 60)]


b = pessoas_ate_60.shape[0]

print(f'Existem {b} pessoas com mais de 51 e atÃ© 60 anos.')
```

```python
pessoas_maiores_61 = dados[(dados['idade_pessoa'] > 61)]


c = pessoas_maiores_61.shape[0]

print(f'Existem {c} pessoas maiores de 61 anos')

```
